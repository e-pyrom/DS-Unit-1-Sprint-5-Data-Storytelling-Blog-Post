```
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```
#data source: https://www.kaggle.com/osmi/mental-health-in-tech-2016

df = pd.read_csv('https://raw.githubusercontent.com/e-pyrom/e-pyrom.github.io/master/mental-health-in-tech-2016_20161114.csv')
```

```
df.head(5)
```

```
#Rename subset columns
df.rename(columns = {'If a mental health issue prompted you to request a medical leave from work, asking for that leave would be:':'Difficulty_for_Taking_MH_Leave',
                   'Have you been diagnosed with a mental health condition by a medical professional?':'Professionally_Diagnosed',
                   'If so, what condition(s) were you diagnosed with?':'Conditions_Diagnosed',
                   'If you have a mental health issue, do you feel that it interferes with your work when being treated effectively?':'Interference_w_Treatment',
                   'If you have a mental health issue, do you feel that it interferes with your work when NOT being treated effectively?': 'Interference_wo_Treatment',
                   'What is your age?':'Age',
                   'What country do you work in?':'Employer_Country',
                   'Which of the following best describes your work position?':'Role_Description',
                   'What US state or territory do you live in?':'State_Terr',
                   'Do you have a family history of mental illness?':'Mental Illness in Family',
                   'Have you had a mental health disorder in the past?':'Mental Illness Past',
                   'Do you currently have a mental health disorder?':'Current Mental Illness',
                   'Do you think that discussing a mental health disorder with your employer would have negative consequences?':'Negative Consequences',
                   'Is your primary role within your company related to tech/IT?':'Role Related to tech/IT',
                   'If yes, what percentage of your work time (time performing primary or secondary job functions) is affected by a mental health issue?':'Time Affected by Mental Health',
                   'Do you believe your productivity is ever affected by a mental health issue?':'Productivity Affected by Mental Health' },
                inplace=True)

df.head(3)
```

```
#Assigning values 

df['Difficulty_for_Taking_MH_Leave'] = df['Difficulty_for_Taking_MH_Leave'].replace(regex='Very easy', value=1)
df['Difficulty_for_Taking_MH_Leave'] = df['Difficulty_for_Taking_MH_Leave'].replace(regex='Somewhat easy', value=2)
df['Difficulty_for_Taking_MH_Leave'] = df['Difficulty_for_Taking_MH_Leave'].replace(regex='Neither easy nor difficult', value=3)
df['Difficulty_for_Taking_MH_Leave'] = df['Difficulty_for_Taking_MH_Leave'].replace(regex='Somewhat difficult', value=4)
df['Difficulty_for_Taking_MH_Leave'] = df['Difficulty_for_Taking_MH_Leave'].replace(regex='Very difficult', value=5)
df['Difficulty_for_Taking_MH_Leave'] = df['Difficulty_for_Taking_MH_Leave'].replace(regex="I don't know", value=0)
df['Difficulty_for_Taking_MH_Leave'].fillna(0, inplace=True)
```

```
df['Negative Consequences'] = df['Negative Consequences'].replace(regex='Yes', value=1)
df['Negative Consequences'] = df['Negative Consequences'].replace(regex='No', value=0)
df['Negative Consequences'] = df['Negative Consequences'].replace(regex='Maybe', value=2)
df['Negative Consequences'].fillna('3', inplace=True)
```

```
df['Productivity Affected by Mental Health'] = df['Productivity Affected by Mental Health'].replace(regex='Yes', value=1)
df['Productivity Affected by Mental Health'] = df['Productivity Affected by Mental Health'].replace(regex='No', value=0)
df['Productivity Affected by Mental Health'] = df['Productivity Affected by Mental Health'].replace(regex='Unsure', value=2)
df['Productivity Affected by Mental Health'] = df['Productivity Affected by Mental Health'].replace(regex='Not applicable to me', value=3)
df['Productivity Affected by Mental Health'].fillna('3', inplace=True)
```

```
#Abbreviating state names

df = df.replace(regex='Alabama', value='AL')
df = df.replace(regex='Alaska', value='AK')
df = df.replace(regex='Arizona', value='AZ')
df = df.replace(regex='California', value='CA')
df = df.replace(regex='Colorado', value='CO')
df = df.replace(regex='Connecticut', value='CT')
df = df.replace(regex='Delaware', value='DE')
df = df.replace(regex='District of Columbia', value='DC')
df = df.replace(regex='Florida', value='FL')
df = df.replace(regex='Georgia', value='GA')
df = df.replace(regex='Idaho', value='ID')
df = df.replace(regex='Illinois', value='IL')
df = df.replace(regex='Indiana', value='IN')
df = df.replace(regex='Iowa', value='IA')
df = df.replace(regex='Kansas', value='KS')
df = df.replace(regex='Kentucky', value='KY')
df = df.replace(regex='Louisiana', value='LA')
df = df.replace(regex='Maine', value='ME')
df = df.replace(regex='Maryland', value='MD')
df = df.replace(regex='Massachusetts', value='MA')
df = df.replace(regex='Michigan', value='MI')
df = df.replace(regex='Minnesota', value='MN')
df = df.replace(regex='Mississippi', value='MS')
df = df.replace(regex='Missouri', value='MO')
df = df.replace(regex='Montana', value='MT')
df = df.replace(regex='Nebraska', value='NE')
df = df.replace(regex='Nevada', value='NV')
df = df.replace(regex='New Hampshire', value='NH')
df = df.replace(regex='New Jersey', value='NJ')
df = df.replace(regex='New Mexico', value='NM')
df = df.replace(regex='New York', value='NY')
df = df.replace(regex='North Carolina', value='NC')
df = df.replace(regex='North Dakota', value='ND')
df = df.replace(regex='Ohio', value='OH')
df = df.replace(regex='Oklahoma', value='OK')
df = df.replace(regex='Oregon', value='OR')
df = df.replace(regex='Pennsylvania', value='PA')
df = df.replace(regex='Rhode Island', value='RI')
df = df.replace(regex='South Carolina', value='SC')
df = df.replace(regex='South Dakota', value='SD')
df = df.replace(regex='Tennessee', value='TN')
df = df.replace(regex='Texas', value='TX')
df = df.replace(regex='Utah', value='UT')
df = df.replace(regex='Vermont', value='VT')
df = df.replace(regex='Virginia', value='VA')
df = df.replace(regex='Washington', value='WA')
df = df.replace(regex='West VA', value='WV')
df = df.replace(regex='Wisconsin', value='WI')
df = df.replace(regex='Wyoming', value='WY')
df = df.replace(regex='Unknown', value='UN')

df['State_Terr'].fillna('UN', inplace=True)
```

```
##SECOND VISUAL##

df['Professionally_Diagnosed'].unique()
```

```
df['disorders'] = df['Professionally_Diagnosed'].str.contains('Yes')
df['disorders']
```

```
condition = df['disorders']==True

disorders = df[condition]
disorders.head(5)
```

```
df['us'] = disorders['Employer_Country'].str.contains('United States of America')

sec_cond = df['us']==True

us_disorders = df[sec_cond]
us_disorders.head(5)
```

```
col = ['State_Terr', 'Age', 'Conditions_Diagnosed', 'Difficulty_for_Taking_MH_Leave', 'Negative Consequences', 'Productivity Affected by Mental Health',
      'Time Affected by Mental Health']
us_cond = us_disorders[col]
us_cond.head(5)
```

```
new_col=['State_Terr', 'Difficulty_for_Taking_MH_Leave', 'Negative Consequences', 'Productivity Affected by Mental Health']
new_df=us_cond[new_col]
new_df.head(3)
```

```
state=new_df.groupby('State_Terr').mean()
state.plot(kind='bar', figsize=(10,8));
plt.title('Leave Difficulty Mental Health')
plt.xlabel('State')
plt.ylabel('Difficulty')
plt.show()
plt.savefig('state.png')
plt.savefig('state.pdf')
```

```
state.describe()
```

```
age_col=['Age', 'Difficulty_for_Taking_MH_Leave', 'Negative Consequences', 'Productivity Affected by Mental Health']
age_df=us_cond[age_col]
age_df.head(3)
```

```
age=age_df.groupby('Age').mean()
age.plot(kind='bar', figsize=(10,8));
plt.title('Leave Difficulty Mental Health')
plt.xlabel('Age')
plt.ylabel('Difficulty')
plt.show()
plt.savefig('age.png')
plt.savefig('age.pdf')
```
